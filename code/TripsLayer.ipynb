```python
import random

import pandas as pd

#读取数据
data = pd.read_csv('../data/sample_taxi.csv')

# data = data[data['is_passenger']==1]
#给数据命名列
data.columns = ['taxi_id', 'time', 'lon', 'lat', 'OpenStatus', 'Speed']
```

```python
data['coordinates'] = data.apply(lambda r: [r['lon'], r['lat']], axis=1)
data['timestamp'] = (
            data['time'].str.slice(0, 2).astype('int') * 360 + data['time'].str.slice(3, 5).astype('int') * 6 + data[
        'time'].str.slice(6, 8).astype('int') * .1).round(2)
data = data.sort_values(by=['taxi_id', 'timestamp'])
data['coordinates'] = data['coordinates'].apply(lambda r: [r])
data['timestamp'] = data['timestamp'].apply(lambda r: [r])
df = data.groupby('taxi_id')['coordinates', 'timestamp'].sum()
df = df.reset_index()
# data = data.sort_values(by = 'timestamp')
# df = pd.DataFrame([[list(data['coordinates'])],[list(data['timestamp'])]]).T
# df.columns = ['coordinates','timestamp']
```
